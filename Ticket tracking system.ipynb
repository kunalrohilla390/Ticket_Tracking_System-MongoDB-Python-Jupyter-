```python
from pymongo import MongoClient
import random

db = MongoClient()["custom_support"]["ticker_tracking_system"]
agents = ["Ravi", "Neha", "Amit", "Priya", "Deepak"]

while True:
    print("1.View Ticket by ID")
    print("2.New Ticket")
    print("3.Help")
    print("4.Exit")
    ch = input("Choice: ")
    if ch == "4": break

    if ch == "2":
        contact = input("Enter Phone OR Address OR Email: ")
        issue = input("Describe Issue: ")
        ticket_id = "TKT" + str(random.randint(1000, 9999))
        assigned = random.choice(agents)

        d = {
            "ticket_id": ticket_id,
            "contact": contact,
            "issue": issue,
            "status": "open",
            "agent": assigned
        }

        db.insert_one(d)
        print(" Ticket Created:", ticket_id)
        print("Assigned to:", assigned)
        print("Agent will contact you within 24 hours.")
        continue

    if ch == "1":# view ticket id
        tid = input("Enter Ticket ID: ")
        t = db.find_one({"ticket_id": tid})
        if not t: print(" Ticket not found"); continue
        print(t)
        if t["status"] == "open":
            print(" Agent", t["agent"], "will contact you within 24 hours.")

    
    if ch == "3":
        contact = input("Enter your Phone OR Email OR Address: ")
        print(" Thank you! Our agent will contact you within 12 hours.")
        issue = input("Please describe your issue: ")
        ticket_id = "TKT" + str(random.randint(1000, 9999))
        assigned = random.choice(agents)

        d = {
            "ticket_id": ticket_id,
            "contact": contact,
            "issue": issue,
            "status": "open",
            "agent": assigned
        }

        db.insert_one(d)
        print(" Ticket Created:", ticket_id)
        print("Assigned to:", assigned)
```
